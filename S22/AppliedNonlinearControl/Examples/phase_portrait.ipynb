### Phase Portraits wih MatPlotLib

In this notebook we will use `streamplot` from `sympy` module to draw the phase portraits of nonlinear system in form: 
\begin{equation*}
\dot{\mathbf{x}} = \mathbf{f}(\mathbf{x})
\end{equation*}

Firstly we define the nonlinear system for which we will draw the phase portrait:

```python
def f(x):
    return x[1], -x[1] - 3*x[0]-x[0]**2
```

Next step is to calculate the grid of points $\mathbf{x} = [x_1, x_2]^T$ and associated $\dot{\mathbf{x}} = [\dot{x}_1, \dot{x}_2]^T$ :

```python
from numpy import linspace, meshgrid

# define the range on which to draw phase portrait
x_range = [5,  # range for x_1
           6]  # range for x_2

x1_max, x2_max = x_range
# define the span for x1 and x2
x1_span = linspace(-1.1*x1_max, 1.1*x1_max, 30)
x2_span = linspace(-1.1*x2_max, 1.1*x2_max, 30)

# Prepare the grid
x1_grid, x2_grid = meshgrid(x1_span, x2_span)

# calculate dxdt for grid points
dx1, dx2 = f([x1_grid, x2_grid])
```

Now we can draw the portrait with help of `matplotlib`:

```python
from matplotlib.pyplot import *

# Setting TeX as output style for text 
rc('text', usetex=True)
# create the figure
figure(figsize=(7, 5)) 
# assign the title
title('Phase Portrait')

# Draw the phase portrait
streamplot(x1_span, x2_span, # span for x1, x2 
           dx1, dx2, # the points calculated on grid 
           arrowsize=1.2, # size of the arrows 
           density=0.9, # density of the vectorfield
           color='k', # color for the trajectories
           linewidth=1, # the width of the lines for trajectories 
           arrowstyle='->') # you can change type of arrows

# Labels and limits for axes
xlabel(r'State  $x_1$')
ylabel(r'State  $x_2$')
xlim([-x1_max, x1_max])
ylim([-x2_max, x2_max])
```
